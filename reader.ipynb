```python
def signal_to_target_df(condition, date=None):
    if date:
        condition1 = condition.loc[date,:]
    else:
        condition1 = condition.iloc[-1, :]
        date = condition.index[-1]
    condition1 = condition1.where(condition1 > 0).dropna()
    condition1 = condition1.reset_index()
    condition1 = condition1.iloc[:, [1, 0]]
    condition1.columns = ['date', 'code']
    condition1['date'] = date
    return condition1

signal_to_target_df(condition)
```

```python
max_price_10 = price.rolling(10).max()
```

```python
def backtest(price, condition):
    buy_price = (price).shift(-1)
    sell_price = (price).shift(-1 - 1)
    return_ = (sell_price / buy_price - 1) * condition
    return_.index = pd.to_datetime(return_.index)
    return_ = (return_.where(return_ != 0).mean(axis=1) - 0.0014).dropna() / 2
    return  return_

return_ = backtest(price, condition)
return_
```

```python
from summarybacktest import SummaryBacktest
s = SummaryBacktest(return_)
s.show_summary()
```

```python
from backtese import Backtest
bt = Backtest()
fac = bt.prc.rolling(10, axis=1).mean()
ic, rankic = bt.ic_info(fac)
turn,ret,sharpe,dd = bt.ret_info(fac)
```
